### i) Consider all stars (within 4kpc) and determine the mean value (and dispersion) of
### the retrieved absolute magnitude. Compare with the true absolute magnitude. Are
### there differences?

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_fwf('Task1.dat', widths = [13, 9, 8, 12, 12, 14, 9, 9, 9], 
               names = ['true parallax (arcsec)', 'true distance (pc)', 'apparent magnitude', 
                        'observed parallax (arcsec)', 'parallax error (arcsec)', 'retrieved distance from observed parallax (pc)',
                        'true absolute magnitude', 'retrieved absolute magnitude', 'photometric error (mag)'
                       ], skiprows = 13)

print (df)

df_within4kpc = df[df['true distance (pc)'] < 4000]
```

```python
print ("Mean retrieved absolute magnitude of stars within 4kpc = %#.4g" % df_within4kpc['retrieved absolute magnitude'].mean())
print ("Dispersion of retrieved absolute magnitude of stars within 4kpc = %#.4g" % df_within4kpc['retrieved absolute magnitude'].std())
```

```python
print ("Mean true absolute magnitude of stars within 4kpc = %#.4g" % df_within4kpc['true absolute magnitude'].mean())
print ("Dispersion of true absolute magnitude of stars within 4kpc = %#.4g" % df_within4kpc['true absolute magnitude'].std())
```

The mean retrieved absolute magnitude is slightly brighter than the true absolute magnitude due to the photometric error introduced into the data - the stars are measured as being further away than they really are, and when the apparent magnitude is converted to absolute magnitude the latter is biased as being too bright.
