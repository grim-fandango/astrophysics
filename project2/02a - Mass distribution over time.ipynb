```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
from os import listdir
from matplotlib import cm

datasets = ['organic', 'gm_late', 'gm_early']

colour_choices=cm.inferno(np.linspace(0, 1, 24))

file_names = np.linspace(0,1, len(datasets))

plt.figure(figsize = (12,10))


    
for dataset in datasets:

    files = listdir('data/' + dataset)
    
    fig, ax = plt.subplots()
    fig.set_size_inches(15,10)
    
    count = 0

    for file in files:
        f = h5py.File('data/' + dataset + '/' + file,'r')

        #create a data table from the HDF5 file
        ds_m = f['Mass']

        # plot histogram from data table
        #fig, ax = plt.subplots()
        colour = colour_choices[count]
        ax.hist(ds_m, color = colour, histtype = 'step', label=dataset)

        #ax.set_xticks([300000, 400000, 500000, 600000, 700000, 800000, 900000, 1000000, 1100000, 1200000, 1300000, 1400000, 1500000, 1600000, 1700000, 1800000])

        
        count = count + 1    
    
    fig.suptitle = ('Organic Masses')
    plt.xlabel('Mass bins ($M_{\odot} x 10^6$)')
    plt.ylabel('Count (logarithmic)')
    plt.yscale('log')
    plt.show()
```
