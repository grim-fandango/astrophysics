```python
import h5py
import matplotlib.pyplot as plt
from os import listdir

files = listdir('data/organic/')

for (file) in files:
    f = h5py.File('data/organic/' + file,'r')
    
    ds_m = f['Mass']

    #list(f.keys())
    #ds_m = f['Mass']
    #ds_m.shape

    #ds_m[0:68]
    print (file)
    
    fig, ax = plt.subplots()
    ax.hist(ds_m, bins=[300000, 400000, 500000, 600000, 700000, 800000, 900000, 1000000, 1100000, 1200000, 1300000, 1400000, 1500000, 1600000, 1700000, 1800000])
    plt.yscale('log')
    ax.set_xticks([300000, 400000, 500000, 600000, 700000, 800000, 900000, 1000000, 1100000, 1200000, 1300000, 1400000, 1500000, 1600000, 1700000, 1800000])
    fig.suptitle = ('Organic Masses')
    plt.show()
```
